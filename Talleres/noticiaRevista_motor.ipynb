```
import requests
import bs4

url = "https://www.motor.com.co/"
response = requests.get(url)
print(response)
print (dir(response))
```

```
print (response.headers)
print(response.headers['Date'])
print(response.text)
```

```
import bs4
completo = bs4.BeautifulSoup(response.text,'html.parser')
noticias = completo.select('.article1x1 a')
print (noticias)
```

```
titulos =[]
links = []
for noticia in noticias:
  titulos.append(noticia.text)
  links.append('https://www.motor.com.co/'+ noticia['href'])
print (titulos)
print (links)
dictNoticias = {}
dictNoticias['Noticias']= titulos
dictNoticias['links']= links
```
